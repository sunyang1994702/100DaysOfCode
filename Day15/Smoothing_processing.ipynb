```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df = df[['date', 'CP1']]
df
```

```python
plt.figure(figsize=(10,10))
plt.plot(df['date'], df['CP1'], label='raw')
plt.xlabel('time')
plt.ylabel('column')
plt.legend()
plt.show()
```

```python
average_span = [10, 20, 40, 60]
```

## ewm smoothing method

```python
average_span_10 = df['CP1'].ewm(span=average_span[0]).mean()
average_span_20 = df['CP1'].ewm(span=average_span[1]).mean()
average_span_40 = df['CP1'].ewm(span=average_span[2]).mean()
average_span_60 = df['CP1'].ewm(span=average_span[3]).mean()
```

```python
plt.figure(figsize=(10,10))
plt.plot(df['date'], df['CP1'], label='raw')
plt.plot(df['date'], average_span_10, label='average_span_10')
plt.plot(df['date'], average_span_20, label='average_span_20')
plt.plot(df['date'], average_span_40, label='average_span_40')
plt.plot(df['date'], average_span_60, label='average_span_60')
plt.xlabel('time')
plt.ylabel('column')
plt.legend()
plt.show()
```

### As the paramater span larger, the smoothing performance better
