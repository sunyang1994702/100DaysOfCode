```python
from multiprocessing import Process
import timeit
```

```python
def sample_func(process_index):
    print('process index: {} started.'.format(process_index))
    num = 0

    for i in range(10000000):
        num += 1
    print('process index: {} ended.'.format(process_index))
```

```python
%%timeit
for i in range(10):
    sample_func(process_index=i)

### 3.83 s ± 42.1 ms per loop (mean ± std. dev. of 7 runs, 1 loop each)
```

```python
%%timeit
process_list = []
for i in range(10):
    process = Process(
        target = sample_func,
        kwargs = {'process_index': i}
    )
    process.start()
    process_list.append(process)
    
for process in process_list:
    process.join()
```
